```python
from transformers import VitsModel, AutoTokenizer
import torch

model = VitsModel.from_pretrained("facebook/mms-tts-mon")
tokenizer = AutoTokenizer.from_pretrained("facebook/mms-tts-mon")
```

```python
import torch
torch.save(model, 'D:/This_season/Tusul/editing_projects/whisper-multiple-hf-datasets-main/model/mms-tts-mon-model.pt')
```

```python
text = "Сайн байна уу намайг отгонбаатар гэдэг."
inputs = tokenizer(text, return_tensors="pt")
with torch.no_grad():
    output = model(**inputs).waveform

print(output.shape)
```

```python
from IPython.display import Audio

Audio(output, rate=model.config.sampling_rate)
```
